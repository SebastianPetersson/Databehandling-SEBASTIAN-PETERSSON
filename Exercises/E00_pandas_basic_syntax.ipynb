````python
kvinnor = pd.read_excel('komtopp50_2020.xlsx', sheet_name='Kvinnor', header=6)
män = pd.read_excel('komtopp50_2020.xlsx', sheet_name='Män', header=6)

def rename_columns(df, kön):
    
    df.columns = ['Rang 2020', 
                  'Rang 2019', 
                  'Kommun', 
                  'Folkmängd 2020', 
                  'Folkmängd 2019', 
                  'Förändring']
    df['Kön'] = kön
    df.drop('Rang 2020', axis = 1, inplace = True)
    df.drop('Rang 2019', axis = 1, inplace = True)

rename_columns(kvinnor, 'Kvinna')
rename_columns(män, 'Man')
````

Första: Här fattade jag uppgiften fel, men visar på den procentuella könsskillnaden, inte specifikt hur många män och kvinnor det faktiskt är.

````python
pivot_df = merged.pivot_table(index = 'Kommun', columns = 'Kön', values = 'Folkmängd 2020')

gender_diff_top_10 = (abs(pivot_df['Man'] - pivot_df['Kvinna']) / (pivot_df['Man'] + pivot_df['Kvinna']) * 100
                      ).rename('Könsskillnad').sort_values(ascending= False).head(10).reset_index()


gender_diff_top_10.plot.bar(
    x = 'Kommun',
    y = 'Könsskillnad',
    color = sns.color_palette('Blues_d', n_colors=len(gender_diff_top_10)),
    title = 'Städerna med störst könsskillnad i Sverige 2020'
    )
plt.ylabel('Procentuell skillnad i %.')
````

Andra och rätta versionen. Var tvungen att ta hjälp för att förstå hur jag skulle kunna plotta datan på det här sättet. \n 
.melt() gör det möjligt att ändra en dataframe från bred format till lång format, vilket i detta fallet behövs för att kunna plotta Man och Kvinna i egna bars.

````python
gender_diff_top_5 = (abs(pivot_df['Man'] - pivot_df['Kvinna']) / (pivot_df['Man'] + pivot_df['Kvinna']) * 100
                      ).rename('Könsskillnad').sort_values(ascending= False).head(5).index
gender_diff_top_5
filtered = pivot_df.loc[gender_diff_top_5]

plot_df = filtered.reset_index().melt(
    id_vars='Kommun',
    value_vars=['Man', 'Kvinna'],
    var_name='Kön',
    value_name='Folkmängd'
)
ax = sns.barplot(
    data = plot_df,
    x = 'Kommun',
    y = 'Folkmängd',
    hue = 'Kön'
)
````
