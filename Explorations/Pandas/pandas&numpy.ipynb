```python
import numpy as np

#ARRAY FUNKTIONEN kommer konvertera lists och tuples till numpy arrays.

sales = [[0, 5, 155, 0, 518], [0, 1827, 616, 317, 325]]

sales_array = np.array(sales)
sales_array

print(f"ndim: {sales_array.ndim}")
print(f"shape: {sales_array.shape}")
print(f"size: {sales_array.size}")
print(f"dtype: {sales_array.dtype}")




```

```python
import numpy as np

rng = np.random.default_rng(616)

prices = (rng.random(10) * 10).round(2)
inventory = rng.integers(20, 100, 10)
inventory[0], inventory [3], inventory[8] = 0, 0, 0
stock = (prices * inventory)
stock.sum().round(1)
```
