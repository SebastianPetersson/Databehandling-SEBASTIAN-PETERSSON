**Syftet med min egna forskning r att frska frutse samband och korrelation i datan, skriva ner en hypotes och sedan pvisa eller motbevisa hypotesen.**  

Jag vill frst utforska korrelationen mellan antalet studenter per lrare (S.F Ratio) och procentandel studenter som tar examen.  

*Hypotes: Om lrare har fler studenter/arbetsbrda br procentandelen studenter som tar examen sjunka.*

```python
display(college['Grad.Rate'].sort_values(ascending=False).head(5), college['Grad.Rate'].sort_values().head(5))
display(college[['Grad.Rate', 'S.F.Ratio']].sort_values('S.F.Ratio', ascending=False))
```

Det visar sig att en av skolorna har en grad rate p 118% vilket inte kan stmma. En gradueringsgrad ver 100% innebr att fler elever tar examen n vad som r registrerade, vilket i praktiken inte r mjligt utan databrister eller fel. Dr av betraktar jag den som en outlier och skapar ett filter som tar bort extremvrden.

```python
filter = college[(college['S.F.Ratio'] < 40) & (college['Grad.Rate'] < 100)]
std_per_teacher = filter[['S.F.Ratio', 'Grad.Rate']].sort_values(by='Grad.Rate', ascending = False)
display(std_per_teacher)
```

```python
from scipy.stats import pearsonr

plt.figure(figsize = (10,8))
sns.lmplot(data=std_per_teacher,
           x = 'S.F.Ratio',
           y = 'Grad.Rate', height = 6, aspect = 1.2)
plt.title('Samband mellan studentlrare-kvot och examensfrekvens')
plt.xlabel('Studenter per lrare (S.F.Ratio)')
plt.ylabel('Examensfrekvens (%)')

correlation = std_per_teacher.corr().loc['S.F.Ratio', 'Grad.Rate']
r_value, p_value = pearsonr(std_per_teacher['S.F.Ratio'], std_per_teacher['Grad.Rate'])
print(f'Korrelation: {correlation:.2f}')
print(f"P-vrde: {p_value}")
```

Datan visar p en relativt tydlig nedgng i examensfrekvens nr S.F.Ratio kar, vilket styrker hypotesen. ven fast korrelationen p -32 r svag, s styrker det lga p-vrdet att sambandet r statistiskt signifikant.

#### **Jag utforskar ven korrelationen mellan 'Personal Spending', 'Expend' (hur mycket skolan betalar fr eleven) och 'Elite'.**

```python
college['Elite_num'] = college['Elite'].map({'Yes': 1, 'No': 0})
subset = college[['Personal', 'Expend', 'Elite']]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(x='Expend', y='Elite', data=subset, ax=axes[0])
axes[0].set_title('Instruktionskostnad vid Elitstatus eller ej')

sns.boxplot(x='Personal', y='Elite', data=subset, ax=axes[1])
axes[1].set_title('Personliga utgifter vid Elitstatus eller ej')

plt.tight_layout()
plt.show()
```

Resultatet ovan r en indikation p att skolor med elitstatus generellt lgger mer pengar p instruktionskostnad n skolor med icke-elitstatus, medan elever har hgre personliga kostnader p icke-elitskolor n elever p skolor med elitstatus. En tolkning p detta skulle kunna vara att skolor med elitstatus lgger mer resurser p undervisning, och att studenter p icke-elitskolor betala mer ur egen ficka. Detta r frsts bara en tolkning som grs utan djupgende forskning, och mnga andra parametrar kan givetvis spela stora faktorer i detta.
