## Uppgift H - Egen utforskning

Jag vill först utforska om det finns någon korrelation mellan antalet studenter per lärare (S.F Ratio) och procentandel studenter som tar examen.  

*Hypotes: Om lärare har fler studenter/arbetsbörda) sjunker procentandelen studenter som tar examen.*

```python
display(college['Grad.Rate'].sort_values(ascending=False).head(5), college['Grad.Rate'].sort_values().head(5))
display(college[['Grad.Rate', 'S.F.Ratio']].sort_values('S.F.Ratio', ascending=False))
```

Det visar sig att en av skolorna har en grad rate på 118% vilket inte kan stämma. En gradueringsgrad över 100% innebär att fler elever tar examen än vad som är registrerade, vilket i praktiken inte är möjligt utan databrister eller fel. Där av betraktar jag den som en outlier och skapar ett filter som tar bort extremvärden.

```python
filter = college[(college['S.F.Ratio'] < 40) & (college['Grad.Rate'] < 100)]
filter.sort_values(by='Grad.Rate', ascending = False)
std_per_teacher = filter[['S.F.Ratio', 'Grad.Rate']]
std_per_teacher
```

```python
plt.figure(figsize = (10,8))
sns.lmplot(data=std_per_teacher,
           x = 'S.F.Ratio',
           y = 'Grad.Rate', height = 6, aspect = 1.2)
plt.title('Samband mellan student–lärare-kvot och examensfrekvens')
plt.xlabel('Studenter per lärare (S.F.Ratio)')
plt.ylabel('Examensfrekvens (%)')

correlation = std_per_teacher.corr().loc['S.F.Ratio', 'Grad.Rate']
print(f'Korrelation: {correlation:.2f}')
```

Datan visar på en relativt tydlig nedgång i examensfrekvens när S.F.Ratio ökar, vilket styrker hypotesen.
